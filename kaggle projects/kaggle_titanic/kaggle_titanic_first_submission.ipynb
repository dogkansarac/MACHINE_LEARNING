```python
# load libraries and data
import pandas as pd
import numpy as np

df = pd.read_csv('train.csv')
df
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
y = df['Survived']
df.drop(['Survived', 'PassengerId','Name','Ticket'], axis=1, inplace = True)
print(y)
df
```

```python
# Imputing numeric missing values (Age)
from sklearn.impute import SimpleImputer 
age = df.iloc[:,2:3]
imputer = SimpleImputer(missing_values = np.nan , strategy = 'median')
age[:] = imputer.fit_transform(age)
print(age.info())

# Imputing caterogical missing values (Cabin, Embarked)
embarked = df.iloc[:,6:8]
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
embarked[:] = imputer.fit_transform(embarked)
print(embarked.info())

# update df
df.loc[:,2:3] = age
df.loc[:,6:8] = embarked
print(df.info())
df
```

```python
for i in range (891):
    df.iloc[i,6] = df.iloc[i,6][0]
df    
```

```python
df = pd.get_dummies(df,columns = ['Sex', 'Cabin', 'Embarked'], drop_first = True)
df
```

```python
from sklearn.model_selection import train_test_split
X_train, X_dev, y_train, y_dev = train_test_split(df, y, test_size = 0.001, random_state = 42)

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)

X_dev = scaler.fit_transform(X_dev)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':(2,3,6,7,8,9,10,12)}
clf = KNeighborsClassifier()
lr = GridSearchCV(clf, parameters)
lr.fit(X_train,y_train)

pred_train = lr.predict(X_train)
pred_dev = lr.predict(X_dev)

from sklearn.metrics import confusion_matrix
cm_dev = confusion_matrix(y_dev, pred_dev)
cm_train = confusion_matrix(y_train, pred_train)
print(cm_train)
print("--------")
print(cm_dev)
print("********")
#from sklearn.metrics import classification_report
print("Train accuracy:", lr.score(X_train, y_train))

print("Dev accuracy:", lr.score(X_dev, y_dev))
```
